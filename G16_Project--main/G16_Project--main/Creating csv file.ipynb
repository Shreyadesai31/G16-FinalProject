```python
import pandas as pd
import os
import numpy as np
from os import *
```

```python
labels=[]
image_name=[]
```

```python

Folder=r"D:/Notes,tut,ppt,videos/Cdac notes/Cdac Essentials/Data/BSRenamed/BSRenamed/BirdSpecies/"
sub_folder= os.listdir(Folder)

     
```

```python
images=pd.DataFrame(image_name)
label=pd.DataFrame(labels)
```
